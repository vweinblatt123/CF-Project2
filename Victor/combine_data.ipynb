```python
import pandas as pd
import numpy as np
from pathlib import Path

%matplotlib inline
```

```python
file_pathS = Path("../Resources/sentiment.csv")
df_sentiment = pd.read_csv(file_pathS, infer_datetime_format = True, parse_dates = True, index_col = "minute")
df_sentiment.sort_index(inplace=True)
```

```python
file_pathD = Path("../WEIRONG/Results/AMC.csv")
df_data = pd.read_csv(file_pathD, infer_datetime_format = True, parse_dates = True, index_col = "date_time")
df_data.sort_index(inplace=True)
```

```python
df_combine = df_data.join(df_sentiment)
```

```python
df_combine = df_combine[df_combine.index > pd.Timestamp("2021-01-28")]
```

```python
file_pathA = Path("../Resources/all_data_csv.csv")
df_macro = pd.read_csv(file_pathA, infer_datetime_format = True, parse_dates = True, index_col = "date_time")
df_macro.sort_index(inplace=True)
```

```python
df_macro = df_macro[['us2y10y', 'vix', 'usdjpy']]
df_combine = df_combine.join(df_macro)
df_combine.fillna(0, inplace = True)
df_combine.to_csv("all_features.csv")
```
